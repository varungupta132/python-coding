```python
import tkinter as tk
from tkinter import *

stringvar = ""
def press(var):
    global stringvar
    stringvar = "You have clicked " + var + " button."
    label2 = Label(root, text = stringvar)
    label2.grid( row=7, column = 1)

if __name__ == "__main__":
    root = tk.Tk()
    root.title("First Example")
    root.geometry("500x500")
    root.configure(background = "cyan")
    
    label1 = Label(root, text = "Your name :: ")
    label1.grid( row=0, column = 0)
    
    entry1 = tk.Entry(root)
    entry1.grid( row=0, column = 1)
    
    listbox1 = tk.Listbox(root)
    listbox1.insert(1, "Python")
    listbox1.insert(2, "Java")
    listbox1.insert(3, "C++")
    listbox1.insert(4, "R")
    listbox1.grid( row=0, column = 2)
    
    var1 = tk.IntVar()
    ckb1 = tk.Checkbutton(root, 
                          text = "Data Structure",
                          variable = var1)
    ckb1.grid(row=1, column = 0)
    var2 = tk.IntVar()
    ckb2 = tk.Checkbutton(root, 
                          text = "Operating System",
                         variable = var2)
    ckb2.grid(row=1, column = 1)
    
   
    button1 = tk.Button(root, 
                        text = "Submit",
                        bg = "light green",
                       activebackground = "pink",
                       height = 1,
                       width = 10,
                       command = lambda: press("Submit"))
    button1.grid( row=2, column = 0)
    button2 = tk.Button(root, 
                        text = "Clear",
                        bg = "light green",
                       activebackground = "pink",
                       height = 1,
                       width = 20,
                       command = lambda: press("Clear"))
    button2.grid( row=3, column = 1)
    
    
    var3 = tk.IntVar()
    rab1 = tk.Radiobutton(root, 
                          text = "Male",
                          variable = var3,
                         value = 1)
    rab1.grid(row=4, column = 0)
    
    rab2 = tk.Radiobutton(root, 
                          text = "Female",
                         variable = var3,
                         value = 2)
    rab2.grid(row=4, column = 1)
    
    var4 = tk.IntVar()
    rab3 = tk.Radiobutton(root, 
                          text = "General",
                          variable = var4,
                         value = 1)
    rab3.grid(row=5, column = 0)
    
    rab4 = tk.Radiobutton(root, 
                          text = "OBC",
                         variable = var4,
                         value = 2)
    rab4.grid(row=5, column = 1)
    
    rab5 = tk.Radiobutton(root, 
                          text = "SC/ST",
                         variable = var4,
                         value = 3)
    rab5.grid(row=5, column = 2)
    
    sb1 = tk.Spinbox(root, from_ = 18, to = 50)
    sb1.grid(row=6, column = 2)
    
    # print(var1)
    # print(var2)
```

```python
from tkinter import *


def pressinc():
    val = label1.cget('text')
    val = int(val) + 5
    label1.config(text = str(val))

def pressdec():
    val = label1.cget('text')
    val = int(val) - 5
    label1.config(text = str(val))

if __name__ == "__main__":
    root = tk.Tk()
    root.title("First Example")
    root.geometry("500x500")
    root.configure(background = "cyan")
    
    myvariable = IntVar()
    
    label1 = Label(root, text = "0")
    label1.grid( row=0, column = 0)
    
    buttoninc = tk.Button(root, 
                        text = "Increment",
                        bg = "light green",
                       activebackground = "pink",
                       height = 1,
                       width = 10,
                       command = pressinc)
    buttoninc.grid( row=1, column = 0)
    
    buttondec = tk.Button(root, 
                        text = "Decrement",
                        bg = "light green",
                       activebackground = "pink",
                       height = 1,
                       width = 10,
                       command = pressdec)
    buttondec.grid( row=2, column = 0)
    
    root.mainloop()
```
