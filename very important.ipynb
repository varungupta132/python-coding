```python
import tkinter as tk
from tkinter import messagebox

def propose():
    girl_name = " xyz"
    if girl_name == " xyz":
        messagebox.showinfo("Proposal", f"I like you, {girl_name}! Will you be my girlfriend?")
    else:
        messagebox.showerror("Error", "Please enter the girl's name!")

# GUI setup
root = tk.Tk()
root.title("Propose to a Girl")

# Heart image
heart_canvas = tk.Canvas(root, width=100, height=100)
# heart_canvas.create_polygon(50, 10, 90, 40, 70, 90, 30, 90, 10, 40, fill="red", outline="")



# Import turtle package 
import turtle 

# Creating a turtle object(pen) 
pen = turtle.Turtle() 

# Defining a method to draw curve 
def curve(): 
	for i in range(200): 

		# Defining step by step curve motion 
		pen.right(1) 
		pen.forward(1) 

# Defining method to draw a full heart 
def heart(): 

	# Set the fill color to red 
	pen.fillcolor('red') 

	# Start filling the color 
	pen.begin_fill() 

	# Draw the left line 
	pen.left(140) 
	pen.forward(113) 

	# Draw the left curve 
	curve() 
	pen.left(120) 

	# Draw the right curve 
	curve() 

	# Draw the right line 
	pen.forward(112) 

	# Ending the filling of the color 
	pen.end_fill() 

# Defining method to write text 
def txt(): 

	# Move turtle to air 
	pen.up() 

	# Move turtle to a given position 
	pen.setpos(-68, 95) 

	# Move the turtle to the ground 
	pen.down() 

	# Set the text color to lightgreen 
	pen.color('lightgreen') 

	# Write the specified text in 
	# specified font style and size 
	pen.write("  I  LIKE   YOU ", font=( 
	"Verdana", 12, "bold")) 


# Draw a heart 
heart() 

# Write text 
txt() 

# To hide turtle 
pen.ht() 




heart_canvas.grid(row=0, column=0, columnspan=2, padx=10, pady=5)

# "I like you" label
like_you_label = tk.Label(root, text="I like you", font=("Arial", 16))
like_you_label.grid(row=1, column=0, columnspan=2, padx=10, pady=5)

# Girl's Name entry
girl_name_label = tk.Label(root, text="Girl's Name:")
girl_name_label.grid(row=2, column=0, padx=10, pady=5, sticky="E")
girl_name_entry = tk.Entry(root)
girl_name_entry.grid(row=2, column=1, padx=10, pady=5, sticky="W")

# Propose Button
propose_button = tk.Button(root, text="Propose", command=propose)
propose_button.grid(row=3, column=0, columnspan=2, padx=10, pady=5)

root.mainloop()
```

```python
import tkinter as tk
from tkinter import messagebox

class TodoList:
    def __init__(self, root):
        self.root = root
        self.root.title("To-Do List")

        self.tasks = []

        self.task_entry = tk.Entry(root, font=('Arial', 14), width=30)
        self.task_entry.grid(row=0, column=0, padx=10, pady=10)

        self.add_button = tk.Button(root, text="Add Task", font=('Arial', 12), command=self.add_task)
        self.add_button.grid(row=0, column=1, padx=5, pady=10)

        self.listbox = tk.Listbox(root, font=('Arial', 12), width=40)
        self.listbox.grid(row=1, column=0, columnspan=2, padx=10, pady=5)

        self.delete_button = tk.Button(root, text="Delete Task", font=('Arial', 12), command=self.delete_task)
        self.delete_button.grid(row=2, column=0, columnspan=2, padx=5, pady=10)

    def add_task(self):
        task = self.task_entry.get()
        if task:
            self.tasks.append(task)
            self.listbox.insert(tk.END, task)
            self.task_entry.delete(0, tk.END)
        else:
            messagebox.showwarning("Warning", "Please enter a task.")

    def delete_task(self):
        try:
            selected_task = self.listbox.curselection()[0]
            self.listbox.delete(selected_task)
            del self.tasks[selected_task]
        except IndexError:
            messagebox.showwarning("Warning", "Please select a task to delete.")

if __name__ == "__main__":
    root = tk.Tk()
    todo_list = TodoList(root)
    root.mainloop()
```
