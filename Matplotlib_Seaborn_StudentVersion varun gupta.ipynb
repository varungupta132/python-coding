**Part 1: Line Plot**
**Question:** Create a line plot showing the sales of a product over a period of 6 months.

**Data:**

1. Months: ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
2. Sales: [150, 200, 250, 300, 350, 400]
**Instructions:**

1. Create a line plot.
2. Add title "Monthly Sales".
3. Label the x-axis as "Months".
4. Label the y-axis as "Sales"
5. Save the plot as "line_plot.png".

**Question:** Customize the line plot by changing the color to green, linestyle to dashed, and adding circle markers.

Instructions:

Customize the plot with the specified styles.
1. Add a legend with the label "Sales".
2. Save the customized plot as "custom_line_plot.png".

```python
# Part 1: Line Plot
# Answer: Create a line plot showing the sales of a product over a period of 6 months.

import matplotlib.pyplot as plt

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [150, 200, 250, 300, 350, 400]

# Create line plot
plt.plot(months, sales)
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.savefig("line_plot.png")
plt.show()

# Answer: Customize the line plot by changing the color to green, linestyle to dashed, and adding circle markers.

# Customize the plot
plt.plot(months, sales, color='green', linestyle='--', marker='o', label='Sales')
plt.title("Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.savefig("custom_line_plot.png")
plt.show()
```

**Part 2: Multiple Line Plots**
**Question:** Create a plot showing the sales for two different products over the same period.

```python
# Part 2: Multiple Line Plots
# Answer: Create a plot showing the sales for two different products over the same period.

# Data for multiple line plot
product_a_sales = [150, 200, 250, 300, 350, 400]
product_b_sales = [100, 150, 200, 250, 300, 350]

# Create multiple line plot
plt.plot(months, product_a_sales, label='Product A Sales')
plt.plot(months, product_b_sales, label='Product B Sales')
plt.title("Product Sales Over 6 Months")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.savefig("multiple_line_plot.png")
plt.show()

# Answer: Customize the multiple line plot by setting different markers and colors for each product.

# Customize the multiple line plot
plt.plot(months, product_a_sales, marker='o', color='blue', linestyle='-', label='Product A Sales')
plt.plot(months, product_b_sales, marker='x', color='red', linestyle='--', label='Product B Sales')
plt.title("Product Sales Over 6 Months")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.savefig("custom_multiple_line_plot.png")
plt.show()
```

**Part 4: Bar Chart**
**Question: **Create a bar chart showing the total sales of different products.

**Data:**

Products: ['Product A', 'Product B', 'Product C']
Sales: [1000, 1500, 1200]
**Instructions:**

1. Create a bar chart with the given data.
2. Add a title "Total Sales by Product".
3. Label the x-axis as "Product".
4. Label the y-axis as "Sales".
5. Save the plot as "bar_chart.png".
**Question:** Customize the bar chart by changing the bar colors and adding value labels on top of each bar.

**Instructions:**

1. Set different colors for each bar.
2. Add value labels on top of each bar to display the sales numbers.
3. Save the customized plot as "custom_bar_chart.png".

```python
# Part 4: Bar Chart
# Answer: Create a bar chart showing the total sales of different products.

# Data for bar chart
products = ['Product A', 'Product B', 'Product C']
sales = [1000, 1500, 1200]

# Create bar chart
plt.bar(products, sales)
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.savefig("bar_chart.png")
plt.show()

# Answer: Customize the bar chart by changing the bar colors and adding value labels on top of each bar.

# Customize the bar chart
colors = ['blue', 'green', 'orange']
bars = plt.bar(products, sales, color=colors)

# Add value labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), va='bottom')

plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Sales")
plt.savefig("custom_bar_chart.png")
plt.show()
```

**Exercise 2: Bar Plot**
1. Load the 'titanic' dataset using Seaborn.

**Exercise 3: Pair Plot**

2. Create a pair plot for all numeric columns.
3. Color the scatter plots based on the 'species' column.
**Exercise 4: Box Plot**
1. Load the 'tips' dataset using Seaborn.
2. Create a box plot showing the distribution of 'total_bill' for each 'day' of the week.
3. Customize the plot by adding appropriate labels and a title.
**Exercise 5: Heatmap**
1. Load the 'flights' dataset using Seaborn.
2. Pivot the data to create a matrix where rows represent 'year' and columns represent 'month'.
3. Create a heatmap showing the number of 'passengers' for each 'year' and 'month'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Exercise 1: Scatter Plot
iris = sns.load_dataset('iris')
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.show()

# Exercise 2: Bar Plot
titanic = sns.load_dataset('titanic')
sns.barplot(data=titanic, x='class', y='fare', errorbar=None)
plt.title('Average Fare by Passenger Class')
plt.ylabel('Average Fare')
plt.xlabel('Passenger Class')
plt.show()

# Exercise 3: Pair Plot
sns.pairplot(iris, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()

# Exercise 4: Box Plot
tips = sns.load_dataset('tips')
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('Box Plot of Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.show()

# Exercise 5: Heatmap
flights = sns.load_dataset('flights')
flights_pivot = flights.pivot_table(index='year', columns='month', values='passengers')
sns.heatmap(flights_pivot, cmap='Blues', annot=True, fmt='.0f')
plt.title('Number of Passengers by Year and Month')
plt.xlabel('Month')
plt.ylabel('Year')
plt.show()
```
